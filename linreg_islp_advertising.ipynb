```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# ordinary least squares multiple linear regression
# predictors X = TV, radio, newspaper
# target Y = sales, using
```

```python
# Importing ISLP Advertising data
df = pd.read_csv("data/islp_advertising.csv")
df.info()
df.describe()
```

```python
# Exploratory data analysis - looking at the covariance matrix of the features

variables = ["TV", "radio", "newspaper", "sales"]

data = np.array(df[variables])

N, p = data.shape

data_normalised = (data - np.mean(data, axis = 0)) / np.std(data, axis = 0)
data_covariance = data_normalised.T @ data_normalised / N

# Printing results

print("N, p = %s, %s\n" % (N, p))

print("data covariance matrix")
print(data_covariance)
```

```python
# Basic sklearn LinearRegression model to check answers against

predictors = ["TV", "radio", "newspaper"]
target = "sales"

X_train = np.array(df[predictors])
Y_train = np.array(df[target])

N, p = X_train.shape

model = LinearRegression()
model.fit(X_train, Y_train)

beta0 = model.intercept_
betap = model.coef_
score = model.score(X_train, Y_train)

# Printing results

print("N, p = %s, %s\n" % (N, p))

print("beta0, betap = %s, %s" % (beta0, betap))
print("Rsquare = %s" % score)
```

```python
# homemade linear regression function

def my_linear_regression(X_train, Y_train):
    
    X_train = np.insert(X_train, 0, 1, axis = 1)
    
    beta = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ Y_train
    
    Yhat = X_train @ beta
    TSS = np.linalg.norm(Y_train - np.mean(Y_train)) ** 2
    RSS = np.linalg.norm(Y_train - Yhat) ** 2
    Rsquare = 1 - RSS / TSS
    
    print("beta = %s" % beta)
    print("Rsquare = %s" % Rsquare)
```

```python
# Testing homemade linear regression function on same data

predictors = ["TV", "radio", "newspaper"]
target = "sales"

X_train = np.array(df[predictors])
Y_train = np.array(df[target])

my_linear_regression(X_train, Y_train)
```

```python
# Simple linear regression - verifying that Rsquare = correlation^2 

variables = ["TV", "radio", "newspaper", "sales"]

data = np.array(df[variables])

N, p = data.shape

data_normalised = (data - np.mean(data, axis = 0)) / np.std(data, axis = 0)
data_covariance = data_normalised.T @ data_normalised / N

# TV and Sales

predictors = ["TV"]
target = "sales"

X_train = np.array(df[predictors])
Y_train = np.array(df[target])

my_linear_regression(X_train, Y_train)

correl_tv_sales = data_covariance[0][3]
print("correl_tv_sales^2 = %s\n" % correl_tv_sales**2)

# radio and Sales

predictors = ["radio"]
target = "sales"

X_train = np.array(df[predictors])
Y_train = np.array(df[target])

my_linear_regression(X_train, Y_train)

correl_radio_sales = data_covariance[1][3]
print("correl_radio_sales^2 = %s\n" % correl_radio_sales**2)

# newspaper and Sales

predictors = ["newspaper"]
target = "sales"

X_train = np.array(df[predictors])
Y_train = np.array(df[target])

my_linear_regression(X_train, Y_train)

correl_newspaper_sales = data_covariance[2][3]
print("correl_newspaper_sales^2 = %s\n" % correl_newspaper_sales**2)
```
